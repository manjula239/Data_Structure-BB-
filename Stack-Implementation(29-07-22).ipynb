```python
class Stack:
    def __init__(self):
        self.st=[]
    def isempty(self):
        return self.st==[]
    def push(self,element):
        self.st.append(element)
    def pop(self):
        if self.isempty():
            return -1
        else:
            return self.st.pop()
    def display(self):
        return self.st
    def peep(self):
        n=len(self.st)
        return self.st[n-1]
    def search(self,element):
        if self.isempty():
            return -1
        else:
            try:
                n=self.st.index(element)
                return len(self.st)-n
            except Valueerror:
                return -2
 ## Ceate empty stack object
s=Stack()
# Menu Driven 
choice=0
while choice<5:
    print("stack operations:")
    print("1:push element")
    print("2:pop element")
    print("3:peep element")
    print("4:search for element")
    print("5:exit")

    choice=int(input("enter your choice:"))
    if choice==1:
        element=int(input("enter element:"))
        s.push(element)
    elif choice==2:
        element=s.pop()
        if element==-1:
            print("The stack is empty")
        else:
            print("poped element is :",element) 
    elif choice==3:
        element=s.peep()
        print("top element:",element)
    elif choice==4:
        element=int(input("enter search element...."))
        pos=s.search(element)
        if pos==-1:
            print("the stack is empty....")
        elif pos==-2:
            print("element not found in stack:")
        else:
            print("element is found at position:",pos)
    else:
        break
        
   ## display the content of stack

    print("stack=",s.display())
```
