##  The process of EDA

```python
# 1.knowing initial details about data
# 2.modifying or removing unwanted data
# 3.Retrieving data
# 4.Getting statistical information
# 5.Drawing graphs/plots
```

```python
import pandas as pd
```

```python
cars=pd.read_csv(r"Downloads\CARS.csv")
cars
```

```python
# To know no of rows and columns are present in the given dataset
print(cars.shape)
```

```python
# To know more information about data like column names and their data types
cars.info()
```

```python
# To display only column names
cars.columns
```

```python
cars.columns.values
```

```python
# To display in the form of list
list(cars.columns)
```

```python
cars.head()
```

```python
cars.tail()
```

```python
cars=cars.drop(['MSRP','Invoice'],axis=1)
cars
```

```python
# To keep only one row and remove all its other duplicates.
cars=cars.drop_duplicates(keep='first')
cars
```

```python
# To know how many rows are having missing values in each column
cars.isnull().sum()
```

```python
# To delete missing value
cars.dropna(inplace=True)
cars
```

```python
# To sort the rows depending on a column
cars_sort=cars.sort_values(by='MPG_City',ascending=False)
cars_sort
```

```python
# Find out how many continets are manufacturing the cars
cars['Origin'].unique()
```

## Getting Statistical Information

```python
cars.describe()
```

```python
# To know the relationship between numeric columns,we can use corr()
cars.corr(method='pearson')
```

```python
# To find correlations between all columns and MPG_City column
cars.corr(method='pearson')['MPG_City']
```

## Drawing Graphs / Plots

```python
# Distribution plot with kernal density estimate (kde) line
import seaborn as sns
sns.displot(data=cars,x='MPG_City',bins=10,kde=True,color='yellow')
```

```python
sns.regplot(x='Length',y='MPG_City',data=cars)
```

```python
cars['Type'].unique()
```

```python
# To know the relationship between the type of the car and mileage in the city.use box plot
box1=sns.boxplot(x='Type',y='MPG_City',data=cars)
```

```python
# To understand relationship between all numeric columns. we can use pairplot()
sns.pairplot(cars,hue='Origin')
```
