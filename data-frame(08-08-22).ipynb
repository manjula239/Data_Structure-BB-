## Advanced Data Analytics using pandas

```python
df1=pd.DataFrame({'player name':['rahul dravid','virat','vinod kamble'],
                 'age':[30,28,22]})
df1                
```

```python
df2=pd.DataFrame({'player name':['rahul dravid','virat','harbajan'],
                 'salary':[45,60,44]})
df2
```

## To join two data frames, we use merge() function

```python
pd.merge(df1,df2, on='player name',how='inner')
```

```python
# left join
pd.merge(df1,df2, on='player name',how='left')
```

```python
# outer join
pd.merge(df1,df2, on='player name',how='outer')
```

```python
df3=pd.DataFrame({'sports person':['rahul dravid','virat','harbajan'],
                 'salary':[45,60,44]})
df3
```

## How to join when there is no common column

```python
pd.merge(df1,df3, left_on='player name',right_on='sports person')
```

```python
pd.merge(df1,df3, left_on='player name',right_on='sports person',how='left')
```

## Concatenation of tables

```python
# All columns of both data frames will be joined side by side
pd.concat([df1,df3])
```

```python
# To ignore existing index values and regenerate them using ignore_index=true
pd.concat([df1,df3],ignore_index=True)
```

```python
# To add rows side by side depending on the column use 'axis=1'
pd.concat([df1,df3],axis=1)
```

## where() method

```python
student={
    "name":['manju','sri','lakshmi','devi'],
    "test1":[15,45,60,76],
    "test2":[45,56,78,80],
    "test3":[89,90,45,40]
}
df=pd.DataFrame(student)
df
```

```python
df.set_index("name",inplace=True)
df
```

```python
# it will display none when condition is true, otherwise it will display first
df.where(df<60,'first')
```

```python
# display first if the marks in the first column (test1) are>=60
df
```

```python
# imitation of where() it will replace first not only test1, other columns, same rows.
df.where(df['test1']<60,'first')
```

```python
df.where(lambda x:x<80,"excellent")
```

## Groupby() method

```python
climate=[('delhi',35,10),
         ('delhi',33,8),
         ('delhi',38,7),
         ('delhi',35,10),
         ('pune',49,15),
         ('pune',50,12),
         ('pune',44,11),
         ('pune',48,13)
        ]
df=pd.DataFrame(climate,columns=["city","tem","wind"])
df
```

```python
df1=df.groupby('city')
df1
```

```python
for city,rows in df1:
    print(city)
    print(rows)
```

```python
df1.get_group('pune')
```

```python
df1.get_group('pune').max()
```

```python
df1.max()
```

```python
df1.agg('max')
```

```python
df1.agg(['max','min','mean'])
```
